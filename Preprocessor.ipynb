```python
import pandas as pd
file_name = 'USC Student Health Website Search Survey (Responses) - Form Responses 1.csv'
colum_names = ['timestamp', 'language', 'contact', 'insurance', 'services', 'emergency', 'covid', 'new_students', 'others']

dataset_df = pd.read_csv(file_name, names = colum_names, skiprows = 1, header = None)
```

```python
dataset_df
```

```python
NaN_val = '[empty]'
dataset_df = dataset_df.fillna(NaN_val)
```

```python
language_col = 'language'
all_col = 'all'

dataset_df[all_col] = dataset_df['contact'].str.cat(dataset_df[colum_names[3:]], sep = ';')
dataset_df
```

```python
dataset_df = dataset_df[[language_col, all_col]]
```

```python
language_groups = dataset_df.groupby(language_col)
```

```python
en = 'English'
es = 'Spanish'
hn = 'Hindi / हिन्दी'
```

```python
import re

def get_queries_for(language):
    lang_df = language_groups.get_group(language)
    all_queries = lang_df[all_col]
    query_list = []
    for query in all_queries:
        unique = set()
        for q in re.split(',|;', query):
            q = str(q.strip())
            if q and q != str(NaN_val):
                unique.add(q)
        query_list = query_list + list(unique)
    return query_list
```

```python
en_queries = get_queries_for(en)
```

```python
es_queries = get_queries_for(es)
```

```python
hn_queries = get_queries_for(hn)
```

```python
import os

current_directory = os.getcwd()
final_directory = os.path.join(current_directory, r'new_dataset')
if not os.path.exists(final_directory):
   os.makedirs(final_directory)
```

```python
def write_to_file(language, queries):
    language = language.split('/')[0]
    file = open('new_dataset/' + language, 'w+')
    for query in queries:
        file.write(query + '\n')
    file.close()
```

```python
write_to_file(en, en_queries)
write_to_file(es, es_queries)
write_to_file(hn, hn_queries)
```
